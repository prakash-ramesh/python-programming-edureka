#String Interpolation - Modulo, .format, f-string, template class

```python
name = "john"
#Modulo
print("Hi % s, Your age is % s" % (name, age))
#.format
print("Hi {}, Your age is {}".format(name, age))
#f-string
print(f"Hi {name}, Your age is {age}")
```

```python
#Template Class
from string import Template
n = Template('Hi $name, your age is $age.')
print(n.substitute(name=name, age=age))
```

#Conditional Statements (if elif else), Loops (while, for else, nested) and Loop Control Statements (pass, continue, break)

```python
#if elif else
if a > int(b):
    print('a>b')
elif int(b) > a:
    print('b>a')
else:
    print('end')
```

```python
#while
count = 0
while count < 5:
    print(count)
    count += 1
```

```python
#while
count = 0
while count < 10:
    count += 1
    if count == 3:
        pass # if loop cant be left empty, dummy stmt
    elif count == 4:
        #4 will not be printed as the continue stmt transfer the control to loop.
        continue
    elif count == 8:
        #break when 8
        break;
    print(count)
```

```python
for i in range(1, 10):
    if i == 3:
        pass # if loop cant be left empty, dummy stmt
    elif i == 4:
        #4 will not be printed as the continue stmt transfer the control to loop.
        continue
    elif i == 8:
        #break when 8
        break;
    print(i)

    for j in range(1,1):
        print(j)
    else:
        print('For Else Loop 2.') #for else will execute when for loop is fully executed.
else:
    #for else wont execute when for loop is breaked.
    print('For Else Loop 1.')
```

#switch/match case

```python
#switch/match case
color = 'green'
match color:
    case 'red':
        print('red')
    case 'yellow':
        print('yellow')
    case _:
        print(color)
```
